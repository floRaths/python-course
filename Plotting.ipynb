```python
import matplotlib.pyplot as plt

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

print(squares)

fig, ax = plt.subplots()
ax.plot(input_values, squares, color ='green', marker='o', linestyle='dashed', linewidth = 2)


# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=20 )
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)
# Set size of tick labels.
ax.tick_params(labelsize=14)

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(11001)
y_values = [x**2 for x in x_values]

figax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.plasma, s = 10)

# Set the range for each axis.
ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style='plain')
plt.savefig('squares_plot.png', bbox_inches='tight')
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
```

```python
dgrey = '0.2'
lgrey = '0.75'

sns.set_style("darkgrid", {'font.family': 'Lato',
                           'text.color': dgrey, # color of non-axes text
                           "axes.facecolor": "white", #color of panel background
                           'axes.edgecolor': lgrey,
                           
                           'axes.grid': False,
                           'grid.color': '0.95',
                           'grid.linestyle': 'solid', #'solid' 'dashed' 'dotted' 'dashdot' 'twodash'
                           
                           'axes.labelsize': '200',
                           'axes.labelcolor': dgrey,
                           'figure.facecolor': 'white',
                           
                           'xtick.color': lgrey,
                           'ytick.color': lgrey,
                    
                           'xtick.bottom': True,
                           'ytick.left': True}
                           )
```

```python
#plt.rcParams['figure.figsize'] = [5, 5]

plot = sns.scatterplot(data = iris, 
                       x    = 'sepal_width', 
                       y    = 'sepal_length',
                       hue  = 'species', 
                       size = 'petal_width',
                       alpha=0.75)

sns.move_legend(plot, "center left", bbox_to_anchor=(1, 0.5), labelspacing=0.25)
sns.set(font_scale=1.2)
        
plt.title('Iris Data Plot', loc = "left", pad = 8, fontdict = {'size' : 'x-large', 'weight' : 'bold'})
plt.tick_params(axis = "both", direction = "out", length = 2.5, pad = 2, labelsize = 10)

#ax = plt.gca()
#for tick_label in ax.get_yticklabels():
#    tick_label.set_color('blue')
```
